###  Data preprocessing

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
import time
from geopy.extra.rate_limiter import RateLimiter
```

*Getting the Station names from the following link*

```python
url='https://www.karnataka.com/bangalore-metro/future-expansion-plans-bangalore-metro-phase-2/'
page=requests.get(url).text
soup=BeautifulSoup(page,'html.parser')
```

```python
station_table=soup.find('table',class_='tablepress tablepress-id-120 tablepress-responsive')
rows= station_table.find_all('tr')
cell=[]
data=[]
```

```python
for i in range(0,len(rows)):
        cell.append(rows[i].find_all('td'))
for x in range(0,38):
    for y in range(0,4):
        if(len(cell[x])>0):
            data.append(cell[x][y].text)
```

```python
Station_names=[]
for dat in data:
    if(len(dat)>0):
        Station_names.append(dat)
```

```python
Stat_name_df=pd.DataFrame({'Station_names':Station_names, 'Latitude': "",'Longitude':""})
Stat_name_df.to_csv('Station_Data.csv')
pd.set_option('display.max_rows',None)
```

### Adding the city name: " ,Bangalore" all the above station name/adressess will enhance the accuracy of co-ordinates of the location while calling Geopy API

```python
Stat_name_df['Station_names']=Stat_name_df['Station_names'].astype(str)+ ", Bangalore"
```

#### Using the Geopy API to obtain the co-ordinates of the location of the Metro Stations

```python
geolocator = Nominatim(user_agent="html")
location=[]
```

```python
x = Stat_name_df['Station_names'].tolist()

for item in x:
    a = geolocator.geocode(item, exactly_one=True, timeout=60)
    try:
        Stat_name_df['Latitude'].iloc[x.index(item)]=a.latitude
    except:
        Stat_name_df['Latitude'].iloc[x.index(item)]=np.nan
        pass
    try:
        Stat_name_df['Longitude'].iloc[x.index(item)]=a.longitude
    except:
        Stat_name_df['Longitude'].iloc[x.index(item)]=np.nan
        pass
    time.sleep(2)
   
```
