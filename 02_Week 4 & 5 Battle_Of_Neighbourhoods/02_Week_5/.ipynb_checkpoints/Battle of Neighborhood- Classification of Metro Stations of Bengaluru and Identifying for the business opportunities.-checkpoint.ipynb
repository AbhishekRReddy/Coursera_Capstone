```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
import time
from geopy.extra.rate_limiter import RateLimiter
import folium
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

*Getting the Station names from the following link*

```python
url='https://www.karnataka.com/bangalore-metro/future-expansion-plans-bangalore-metro-phase-2/'
page=requests.get(url).text
soup=BeautifulSoup(page,'html.parser')
```

```python
station_table=soup.find('table',class_='tablepress tablepress-id-120 tablepress-responsive')
rows= station_table.find_all('tr')
cell=[]
data=[]
```

```python
for i in range(0,len(rows)):
        cell.append(rows[i].find_all('td'))
for x in range(0,38):
    for y in range(0,4):
        if(len(cell[x])>0):
            data.append(cell[x][y].text)
```

```python
Station_names=[]
for dat in data:
    if(len(dat)>0):
        Station_names.append(dat)
```

```python
Stat_name_df=pd.DataFrame({'Station_names':Station_names, 'Latitude': "",'Longitude':""})
Stat_name_df.to_csv('Station_Data.csv')
pd.set_option('display.max_rows',None)
```

### Adding the city name: " ,Bangalore" all the above station name/adressess will enhance the accuracy of co-ordinates of the location while calling Geopy API

```python
Stat_name_df['Station_names']=Stat_name_df['Station_names'].astype(str)+ ", Bangalore"
```

#### Using the Geopy API to obtain the co-ordinates of the location of the Metro Stations

```python
geolocator = Nominatim(user_agent="html")
location=[]
```

```python
x = Stat_name_df['Station_names'].tolist()

for item in x:
    a = geolocator.geocode(item, exactly_one=True, timeout=60)
    try:
        Stat_name_df['Latitude'].iloc[x.index(item)]=a.latitude
    except:
        Stat_name_df['Latitude'].iloc[x.index(item)]=np.nan
        pass
    try:
        Stat_name_df['Longitude'].iloc[x.index(item)]=a.longitude
    except:
        Stat_name_df['Longitude'].iloc[x.index(item)]=np.nan
        pass
    time.sleep(2)
   
```

```python
categories_list = []
# Let's print only the top-level categories and their IDs and also add them to categories_list

def print_categories(categories, level=0, max_level=0):    
    if level>max_level: return
    out = ''
    out += '-'*level
    for category in categories:
        print(out + category['name'] + ' (' + category['id'] + ')')
        print_categories(category['categories'], level+1, max_level)
        categories_list.append((category['name'], category['id']))
        
print_categories(results['response']['categories'], 0, 0)
```

### K-means Clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### From the above elbow method we can choose the number of cluster as 4

```python
# set number of clusters
kclusters = 4
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(cluster_df)

kmeans_labels = kmeans.labels_
# Change label numbers so they go from highest scores to lowest
replace_labels = {0:2, 1:0, 2:3, 3:1}
for i in range(len(kmeans_labels)):
    kmeans_labels[i] = replace_labels[kmeans_labels[i]]

stations_clusters_df = venues.copy()
stations_clusters_df['Cluster'] = kmeans_labels
stations_clusters_minmax_df = cluster_df.copy()
stations_clusters_minmax_df['Cluster'] = kmeans_labels
stations_clusters_minmax_df['English name'] = stations_venues_df['Station_names']
stations_clusters_minmax_df['Coordinates'] = stations_venues_df['Coordinates']
```

```python
import matplotlib.ticker as ticker

fig, axes = plt.subplots(1,kclusters, figsize=(20, 10), sharey=True)

axes[0].set_ylabel('Count of venues (relative)', fontsize=25)
#plt.set_xlabel('Venue category', fontsize='x-large')

for k in range(kclusters):
    #Set same y axis limits
    axes[k].set_ylim(0,1.1)
    axes[k].xaxis.set_label_position('top')
    axes[k].set_xlabel('Cluster ' + str(k), fontsize=25)
    axes[k].tick_params(labelsize=20)
    plt.sca(axes[k])
    plt.xticks(rotation='vertical')
    sns.boxplot(data = stations_clusters_minmax_df[stations_clusters_minmax_df['Cluster'] == k].drop('Cluster',1), ax=axes[k])

plt.show()
```

```python
#create map of Moscow with all stations
map_bengaluru = folium.Map(location=[12.9716, 77.5946], zoom_start=11)



#add markers
for i, station, coordinates, cluster in zip(stations_clusters_minmax_df.index,
                                         stations_clusters_minmax_df['English name'],
                                         stations_clusters_minmax_df['Coordinates'],
                                         stations_clusters_minmax_df['Cluster']):
    latlong = [float(x) for x in coordinates.split(',')]    
    #colors = sns.color_palette(None, kclusters).as_hex()
    colors=['blue','black','orange','red']
    #Calculate top 3 categories for each station
    station_series = stations_clusters_minmax_df.iloc[i]
    top_categories_dict = {}
    for cat in categories_list:
        top_categories_dict[cat[0]] = station_series[cat[0]]
    top_categories = sorted(top_categories_dict.items(), key = lambda x: x[1], reverse=True)
    popup='<b>{}</b><br>Cluster {}<br>1. {} {}<br>2. {} {}<br>3. {} {}'.format(
        station,
        cluster,
        top_categories[0][0],
        "{0:.2f}".format(top_categories[0][1]),
        top_categories[1][0],
        "{0:.2f}".format(top_categories[1][1]),
        top_categories[2][0],
        "{0:.2f}".format(top_categories[2][1]))
    folium.CircleMarker(
        latlong,
        fill=True,
        fill_opacity=0.5,
        popup=folium.Popup(popup, max_width = 300),
        radius=5,
        color=colors[cluster]
    ).add_to(map_bengaluru)

map_bengaluru
```
